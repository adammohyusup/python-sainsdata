## Visualisasi Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('covid_19_indonesia_time_series_all.csv')

x_var = 'Location'
y_var = 'Total Cases'

data_filtered = data[data['Location'] != 'Indonesia']

data_filtered = data_filtered.sort_values(by=y_var, ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=x_var, y=y_var, data=data_filtered, palette='Spectral', hue=x_var, dodge=False, legend=False)

# Customize the chart
plt.title(f'Total Cases by {x_var} (excluding Indonesia)')
plt.xlabel(x_var)
plt.ylabel(y_var)
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('covid_19_indonesia_time_series_all.csv')

data_filtered = data[data['Location'] != 'Indonesia']

x_var = 'Location'  

y_var = 'New Cases'  

plt.figure(figsize=(12, 6))
sns.barplot(x=x_var, y=y_var, data=data_filtered, palette='viridis', hue=x_var, dodge=False, legend=False)

plt.title('New Cases by Location')
plt.xlabel(x_var)
plt.ylabel(y_var)
plt.xticks(rotation=45, ha='right')  

plt.tight_layout()  

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('covid_19_indonesia_time_series_all.csv')

data_filtered = data[data['Location'] != 'Indonesia']

total_cases_by_location = data_filtered.groupby('Location')['Total Cases'].max()

total_cases_all_locations = total_cases_by_location.sum()

percentage_total_cases = (total_cases_by_location / total_cases_all_locations) * 100

plt.figure(figsize=(10, 10))
plt.pie(percentage_total_cases, labels=percentage_total_cases.index, autopct="%1.1f%%", labeldistance=1.1)
plt.title("Percentage of Total Cases by Location")

plt.axis('equal')  

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('covid_19_indonesia_time_series_all.csv')

x_axis = "Date"
y_axis = "Total Cases"

data_filtered = data[data['Location'] != 'Indonesia']

grouped_data = data_filtered.groupby("Location")[y_axis]

plt.figure(figsize=(10, 6))
for name, group_data in grouped_data:
    plt.plot(group_data.index, group_data.values, label=name)

plt.title(f"{y_axis} over {x_axis} (excluding Indonesia)")
plt.xlabel(x_axis)
plt.ylabel(y_axis)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 
plt.grid(True)
plt.tight_layout()  

plt.show()
```

## GIS

```python
import numpy as np
import pandas as pd
import geopandas as gpd
import numpy as np
import seaborn as sns
import textwrap
import matplotlib.pyplot as plt
import datetime
from datetime import date
```

```python
df = pd.read_csv("covid_19_indonesia_time_series_all.csv")
covid_df = df.copy()
covid_df.head()
```

```python
date2 =[]
for item in covid_df['Date']:
    item2=item.split('/')
    month=int(item2[0])
    day=int(item2[1])
    year=int(item2[2])
    date2+=[datetime.date(year,month,day)]
covid_df['Date']=date2
covid_df['Date'] = pd.to_datetime(df['Date'])
covid_df.head()
```

```python
covid_df.drop(["Location ISO Code", "New Cases", "New Deaths", "New Recovered", "New Active Cases", "Province", "Country", "Continent", "Island", "Time Zone", 
                "Special Status", "Total Regencies", "Total Cities", "Total Districts", "Total Urban Villages", "Total Rural Villages", "Area (km2)", 
                "New Cases per Million", "Total Cases per Million","New Deaths per Million", "Total Deaths per Million", "Case Fatality Rate",
                "Case Recovered Rate", "Growth Factor of New Cases","Growth Factor of New Deaths", "City or Regency"],axis=1,inplace=True)
```

```python
covid_newest = covid_df.drop_duplicates(subset='Location', keep="last").copy()
covid_newest.head()
```

```python
data_geo = gpd.read_file('gadm36_IDN_3.shp')
data_geo.drop(["NAME_0"],axis=1,inplace=True)
data_geo.rename(columns = {"NAME_1":"Provinsi"}, inplace = True)
data_geo['Provinsi'].replace({"Jakarta Raya":"DKI Jakarta"},inplace=True)
data_geo.head()
```

```python
covid_newest.rename(columns={'Location':'Provinsi'},inplace=True)
covid_newest
```

```python
covid_peta = data_geo.merge(covid_newest.sort_values(by='Provinsi'),on='Provinsi')
covid_peta
```

```python
fig,ax = plt.subplots(4,figsize=(36,36))
plt.style.use('ggplot')
covid_peta.plot(column='Total Cases',cmap='Reds',figsize=(9,36),legend=True,ax=ax[0])
ax[0].set_title('Total kasus',fontsize=25)
covid_peta.plot(column='Total Deaths',cmap='Purples',figsize=(9,36),legend=True,ax=ax[1])
ax[1].set_title('Total Kematian',fontsize=25)
covid_peta.plot(column='Total Recovered',cmap='Greens',figsize=(9,36),legend=True,ax=ax[2])
ax[2].set_title('Total Sembuh',fontsize=25)
covid_peta.plot(column='Total Active Cases',cmap='Blues',figsize=(9,36),legend=True,ax=ax[3])
ax[3].set_title('Total Kasus Aktif',fontsize=25)

#add background color
ax[0].set_facecolor('#e0e0e0')
ax[1].set_facecolor('#e0e0e0')
ax[2].set_facecolor('#e0e0e0')
ax[3].set_facecolor('#e0e0e0')

for i in np.arange(4):
    ax[i].spines['top'].set_visible(False)
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['left'].set_visible(False)
    ax[i].spines['bottom'].set_visible(False)
    for item in ax[i].get_xticklabels():
        item.set_visible(False)
    for item in ax[i].get_yticklabels():
        item.set_visible(False)
    ax[i].set_xticks([])
    ax[i].set_yticks([])
fig.suptitle('Peta Covid-19 Indonesia', fontsize=18)
```
